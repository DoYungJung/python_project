```python
todayDate = dt.datetime.now().date()

time = todayDate

# print(시작시간으로부터 8일뒤)

# time + timedelta(days =-8)

needtime = time + timedelta(days =-8)

preDate = str(needtime)


preDate.split("-")

preYear = preDate[:4]

preWeek = preDate[5:7]

preDay = preDate[-2:]

startDate = preYear + preWeek + preDay

# print(시작시간으로부터 1일뒤)

needEndTime = time + timedelta(days =-1)

preEndDate = str(needEndTime)

preEndDate.split("-")

preEndYear = preEndDate[:4]

preEndWeek = preEndDate[5:7]

preEndDay = preEndDate[-2:]

endDate = preEndYear + preEndWeek + preEndDay

mainUrl = "http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList"

dataFramesList = []
```

## 5.정보추출하기

```python
for i in range(0,stnListLen):
    queryParams = '?' + urlencode(
         {
                quote_plus('ServiceKey') :serviceKey0,
                quote_plus('numOfRows'): "20",
                quote_plus('pageNo'): "10",
                quote_plus('dataCd') :"ASOS",
                quote_plus('dateCd') :"DAY",
                quote_plus('startDt') : startDate,
                quote_plus('endDt') : endDate,
                quote_plus('stnIds'): stnList[i]
         }
    )
    
    targetUrl = mainUrl + queryParams
#   타겟url 만들어서 접근
    response = requests.get(targetUrl)
    response.encoding = "utf-8"
    htmlNatureCode = response.text

    html = bs4.BeautifulSoup(htmlNatureCode, "lxml-xml")
    bs = html
    itmesTag=bs.find(name="items")
    rows=itmesTag.findAll(name="item")
#   태그별 접근
    rowsLen = len(rows)

    nameList =[]
    columnsList =[]
    rowsList = []
#   정보를 담기 위한 리스트 생성
    for j in range(0,rowsLen):
        columns=rows[j].findAll()
        for k in range(0,len(columns)):
            if(columns[k].text == ''):
                columnsList.append("none")
            else:
                columnsList.append(columns[k].text)
        rowsList.append(columnsList)
        columnsList = []

    eachRows = rows[0].findAll()

    rowsLen = len(rows[0])

    for j in range(0,rowsLen):
        nameList.append(eachRows[j].name)

    dataFrame=pd.DataFrame(rowsList, columns = nameList)
    dataFramesList.append(dataFrame)    

totalFrame = pd.concat(dataFramesList, ignore_index = True)
# 지점별로 나오는 정보를 합치기 위하여 concat 활용
totalFrame.to_csv("{}-{}.csv".format(k,endDate),mode="w", encoding = "ms949")
```
